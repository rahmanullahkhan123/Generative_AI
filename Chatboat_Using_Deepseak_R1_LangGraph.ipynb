```python
from typing import TypedDict, List, Dict

from langgraph.graph import StateGraph, END
from langchain_ollama import OllamaLLM
```

```python
class ChatState(TypedDict):
    question: str
    answer: str
    history: List[Dict[str, str]]
```

```python
[
    {
        "user": "What is AI?",
        "assistant": "AI is..."
    }
]
```

```python
llm = OllamaLLM(
    model="deepseek-r1:1.5b",
    temperature=0.7
)
```

```python
def chatbot(state: ChatState):

    question = state["question"]

    response = llm.invoke(question)

    state["answer"] = response


    # Save conversation
    state["history"].append(
        {
            "user": question,
            "assistant": response
        }
    )

    return state
```

```python
graph = StateGraph(ChatState)


# Add node
graph.add_node(
    "chatbot",
    chatbot
)


# Start point
graph.set_entry_point(
    "chatbot"
)


# End point
graph.add_edge(
    "chatbot",
    END
)


# Compile
chatbot_graph = graph.compile()
```

```python
history = []


def chat():

    print("🤖 DeepSeek-R1 Local Chatbot")
    print("Type 'exit' or 'stop' to close\n")


    while True:

        user_input = input("You: ")


        # Stop chatbot
        if user_input.lower() in [
            "exit",
            "stop",
            "quit"
        ]:
            print("Bot: Goodbye 👋")
            break



        result = chatbot_graph.invoke(
            {
                "question": user_input,
                "answer": "",
                "history": history
            }
        )


        print(
            "\nBot:",
            result["answer"]
        )


        print("-"*50)
```

```python
chat()
```
