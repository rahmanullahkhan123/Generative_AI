```python
from typing import TypedDict, Annotated

from langgraph.graph import StateGraph, END

from langchain_ollama import ChatOllama

from langchain_core.messages import HumanMessage

from langchain_community.tools import DuckDuckGoSearchRun
```

```python
llm = ChatOllama(
    model="llama3.2",
    temperature=0.3
)
```

```python
class AgentState(TypedDict):

    user_input:str

    task_result:str

    research_result:str

    email_result:str

    final_answer:str
```

```python
def task_agent(state):

    prompt=f"""

You are a productivity planner.

Create a detailed plan for:

{state['user_input']}

Provide steps and timeline.

"""

    response=llm.invoke(prompt)


    state["task_result"]=response.content


    return state
```

```python
search = DuckDuckGoSearchRun()
```

```python
def research_agent(state):


    query=state["user_input"]


    result=search.run(query)


    state["research_result"]=result


    return state
```

```python
search = DuckDuckGoSearchRun()
```

```python
def research_agent(state):


    query=state["user_input"]


    result=search.run(query)


    state["research_result"]=result


    return state
```

```python
def email_agent(state):


    prompt=f"""

Write a professional email based on:

{state['user_input']}


"""


    response=llm.invoke(prompt)


    state["email_result"]=response.content


    return state
```

```python
def supervisor(state):


    query=state["user_input"].lower()


    if "plan" in query or "learn" in query:

        return "task"


    elif "search" in query or "research" in query:

        return "research"


    elif "email" in query:

        return "email"


    else:

        return "task"
```

```python
workflow = StateGraph(AgentState)
```

```python
workflow.add_node(
    "task",
    task_agent
)


workflow.add_node(
    "research",
    research_agent
)


workflow.add_node(
    "email",
    email_agent
)
```

```python
workflow.set_conditional_entry_point(
    supervisor,
    {
        "task":"task",
        "research":"research",
        "email":"email"
    }
)

```

```python
workflow.add_edge(
    "task",
    END
)


workflow.add_edge(
    "research",
    END
)


workflow.add_edge(
    "email",
    END
)
```

```python
app = workflow.compile()
```

```python
result = app.invoke(
{
"user_input":
"Create a plan to learn Generative AI"
}
)
```

```python
print(result["task_result"])
```

```python
result=app.invoke(
{
"user_input":
"Research latest AI trends"
}
)


print(result["research_result"])
```
